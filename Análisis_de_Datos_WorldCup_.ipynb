
Goleadores Históricos

```python
df_players.head()
```

```python
df_players_event= df_players.dropna(subset= ["Event"])
df_players_event
```

```python
def cantidad_goles(cadena):
  return cadena.count("G")
```

```python
df_players_event["Goals"]= df_players_event["Event"].apply(cantidad_goles)
```

```python
df_players_event.head()
```

```python
df_players_event.get(["Player Name", "Goals"]).groupby("Player Name").sum().sort_values(by="Goals", ascending= False).head(10)
```
